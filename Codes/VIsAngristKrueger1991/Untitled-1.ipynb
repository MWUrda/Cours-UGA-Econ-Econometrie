```python
import pandas as pd 
import statsmodels.api as sm
```

```python
ak1991df = pd.read_excel('https://www.ssc.wisc.edu/~bhansen/econometrics/AK1991.xlsx')
ak1991df.info()
```

```python
yob_region_dummies = pd.get_dummies(data = ak1991df[['yob', 'region']], columns= ['yob', 'region'])
yob_region_dummies.info()
ak1991df2 = pd.concat([ak1991df,  yob_region_dummies], axis=1)
```

```python
ak1991df = pd.concat([ak1991df,  yob_region_dummies], axis=1)
print(ak1991df.shape)
print(ak1991df.info())
ak1991df.head()
```

```python
Y = ak1991df['logwage']
X = ak1991df[['edu', 'black', 'smsa', 'married', 'region_1', 'region_2', 'region_3', 'region_4', 'region_5',
'region_6', 'region_7', 'region_8', 'yob_1931', 'yob_1932', 'yob_1933', 'yob_1934', 'yob_1935', 'yob_1936',
'yob_1937', 'yob_1938', 'yob_1939']]
```

```python
import statsmodels.api as sm
model = sm.OLS(Y, sm.add_constant(X))
results = model.fit()
print(results.summary())
```
